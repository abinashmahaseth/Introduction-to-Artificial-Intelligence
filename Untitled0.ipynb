```python
# Pixel pushing in NumPy
# This notebook demonstrates basic image manipulation with NumPy and Matplotlib.
# It introduces NumPy arrays, image visualization, and mathematical patterns.

# Import required libraries
from matplotlib import pyplot as plt
import numpy as np
import math

# Cell 1: Random 10x10 grayscale image
# Create a 10x10 array with random values in [0, 1]
my_array = np.random.random((10, 10))
print(my_array)

# Display as grayscale image, ensuring values are in [0, 1]
plt.imshow(my_array, cmap='gray', clim=(0, 1), interpolation='nearest')
plt.title('Random 10x10 Image')
plt.colorbar()
plt.show()

# Cell 2: Single pixel image
# Create a 10x10 zero array and set one pixel to 1
myImage = np.zeros((10, 10))
myImage[1, 0] = 1

# Display the image
plt.imshow(myImage, cmap='gray', clim=(0, 1), interpolation='nearest')
plt.title('Single Pixel at (1, 0)')
plt.colorbar()
plt.show()

# Cell 3: Drawing a smooth circle
# Create a higher-resolution 240x320 array for better quality
myImage = np.zeros((240, 320))

# Parameters for the circle
center_x, center_y = 160, 120  # Center of the image
radius = 30
segments = 600
TWO_PI = 2 * np.pi
spacing = TWO_PI / segments

# Use vectorized operations to compute circle points
theta = np.linspace(0, TWO_PI, segments, endpoint=False)
x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

# Anti-aliasing: Blend pixel values based on sub-pixel coverage
for xi, yi in zip(x, y):
    x_floor, y_floor = int(xi), int(yi)
    x_frac, y_frac = xi - x_floor, yi - y_floor

    # Distribute intensity to four neighboring pixels
    myImage[y_floor, x_floor] += (1 - x_frac) * (1 - y_frac)
    myImage[y_floor, x_floor + 1] += x_frac * (1 - y_frac) if x_floor + 1 < 320 else 0
    myImage[y_floor + 1, x_floor] += (1 - x_frac) * y_frac if y_floor + 1 < 240 else 0
    myImage[y_floor + 1, x_floor + 1] += x_frac * y_frac if x_floor + 1 < 320 and y_floor + 1 < 240 else 0

# Normalize to [0, 1]
myImage = np.clip(myImage, 0, 1)

# Display with bilinear interpolation for smoothness
plt.imshow(myImage, cmap='gray', clim=(0, 1), interpolation='bilinear')
plt.title('Smooth Circle (Radius = 30)')
plt.colorbar()
plt.show()

# Cell 4: Radial sine wave pattern
# Create a 240x320 array
myImage = np.zeros((240, 320))

# Vectorized computation of radial distance
i, j = np.indices((240, 320))
distance = np.sqrt((i - 120)**2 + (j - 160)**2)  # Center at (120, 160)
frequency = 3 * np.pi / 240  # Adjusted for visible waves

# Compute sine pattern and normalize to [0, 1]
myImage = np.sin(distance * frequency)
myImage = (myImage + 1) / 2  # Map [-1, 1] to [0, 1]

# Display the pattern
plt.imshow(myImage, cmap='gray', clim=(0, 1), interpolation='bilinear')
plt.title('Radial Sine Wave Pattern')
plt.colorbar()
plt.show()

# Cell 5: Tangent and cosine pattern
# Create a 240x320 array
myImage = np.zeros((240, 320))

# Vectorized computation
i, j = np.indices((240, 320))
frequency = np.pi / 320 * 10  # Reduced frequency for smoother patterns
t = (np.tan(i * frequency) * np.cos(j * frequency) +
     np.arctan(j * frequency) * np.cos(i * frequency))

# Normalize to [0, 1]
t_min, t_max = t.min(), t.max()
myImage = (t - t_min) / (t_max - t_min) if t_max != t_min else t

# Display the pattern
plt.imshow(myImage, cmap='gray', clim=(0, 1), interpolation='bilinear')
plt.title('Tangent and Cosine Pattern')
plt.colorbar()
plt.show()
```

```python
# Cell 6: Custom grid pattern with sine and cosine waves
import numpy as np
import matplotlib.pyplot as plt

# Create a 240x320 array
myImage = np.zeros((240, 320))

# Vectorized computation
i, j = np.indices((240, 320))
f = 20.0  # Frequency for wave spacing

# Compute the custom expression
t = np.sin(i / f) + np.cos(j / f)

# Normalize to [0, 1]
myImage = (t + 2) / 4

# Display the pattern
plt.imshow(myImage, cmap='gray', clim=(0, 1), interpolation='bilinear')
plt.title('Custom Sine-Cosine Grid Pattern')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Create a 240x320 grayscale image
myImage = np.zeros((240, 320))

# Generate pixel values using the custom expression
for i in range(240):
    for j in range(320):
        t = math.sin(i / 15) * math.cos(j / 25) + math.exp(-((i-120)**2 + (j-160)**2) / 8000)
        myImage[i, j] = t

# Normalize the image to [0, 1] for proper display
min_val = np.min(myImage)
max_val = np.max(myImage)
if max_val > min_val:  # Avoid division by zero
    myImage = (myImage - min_val) / (max_val - min_val)

# Display the image
plt.imshow(myImage, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()

# Optionally, save the image
plt.imsave('custom_image.png', myImage, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Image dimensions
width, height = 256, 256

# Create a 2D array for a grayscale image
image = np.zeros((height, width))

# Generate pixel values using a custom expression
for y in range(height):
    for x in range(width):
        # Normalize coordinates to [-1, 1]
        xn = 2 * x / width - 1
        yn = 2 * y / height - 1
        # Custom expression: spiral pattern
        value = np.sin(10 * np.arctan2(yn, xn) + 5 * np.sqrt(xn**2 + yn**2))
        # Normalize to [0, 1]
        image[y, x] = (value + 1) / 2

# Display the image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
