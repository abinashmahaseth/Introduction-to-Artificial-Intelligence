```python
# drawing a circle
# looks terrible doesn't it?

# so numpy arrays are column major, not row major.
# So you need to do height first
myImage = np.zeros((240,320))

# This code is exactly the same as the the code frmo
TWO_PI = 3.14159 * 2
segments = 600
spacing = TWO_PI / segments
size = 30

for i in range(segments):
    x = math.cos(spacing * i) * size
    y = math.sin(spacing * i) * size
    myImage[math.floor(y) + 50, math.floor(x) + 50] = 1 # notice we need to floor the output to ints

plt.imshow(myImage, interpolation="bilinear", clim=(0,1),cmap="gray")
plt.show()
```

```python
myImage = np.zeros((240,320))

# This code is exactly the same as the the code from our JS and C++ examples

centre = 100
size = 50

for i in range(240):
    for j in range(320):
        x_dist = abs(centre-i)
        y_dist = abs(centre-j)
        dist = math.sqrt(x_dist * x_dist + y_dist * y_dist)
        if dist < size:
            myImage[i,j] = 1

#plt.imshow(myImage, clim=(0,1),cmap="gray")
plt.imshow(myImage, interpolation="bilinear", clim=(0,1), cmap="gray")
plt.show()
```

```python
myImage = np.zeros((240,320))

# uses numpy distance measures but is lots slower..

centre = 100
size = 50

for i in range(240):
    for j in range(320):
        positions=np.array([[i,j],[abs(centre-i),abs(centre-j)]])
        distance = np.linalg.norm(positions[0:1] - positions[1:2])
        if distance < size:
            myImage[i,j] = 1

plt.imshow(myImage, clim=(0,1),cmap="gray")
plt.imshow(myImage, interpolation="bilinear", clim=(0,1), cmap="gray")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Image dimensions
width, height = 256, 256

# Create a 2D array for a grayscale image
image = np.zeros((height, width))

# Generate pixel values using a custom expression
for y in range(height):
    for x in range(width):
        # Normalize coordinates to [-1, 1]
        xn = 2 * x / width - 1
        yn = 2 * y / height - 1
        # Custom expression: spiral pattern
        value = np.sin(10 * np.arctan2(yn, xn) + 5 * np.sqrt(xn**2 + yn**2))
        # Normalize to [0, 1]
        image[y, x] = (value + 1) / 2

# Display the image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Create a 240x320 grayscale image
myImage = np.zeros((240, 320))

# Generate pixel values using the custom expression
for i in range(240):
    for j in range(320):
        t = math.sin(i / 15) * math.cos(j / 25) + math.exp(-((i-120)**2 + (j-160)**2) / 8000)
        myImage[i, j] = t

# Normalize the image to [0, 1] for proper display
min_val = np.min(myImage)
max_val = np.max(myImage)
if max_val > min_val:  # Avoid division by zero
    myImage = (myImage - min_val) / (max_val - min_val)

# Display the image
plt.imshow(myImage, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()

# Optionally, save the image
plt.imsave('custom_image.png', myImage, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Create a 240x320 grayscale image
myImage = np.zeros((240, 320))

# Generate pixel values using a radial wave expression
for i in range(240):
    for j in range(320):
        # Distance from center (120, 160)
        r = math.sqrt((i - 120)**2 + (j - 160)**2)
        t = math.sin(r / 10) * math.cos(r / 20) + 0.5  # Radial waves
        myImage[i, j] = t

# Normalize the image to [0, 1]
min_val = np.min(myImage)
max_val = np.max(myImage)
if max_val > min_val:
    myImage = (myImage - min_val) / (max_val - min_val)

# Display the image
plt.imshow(myImage, cmap='gray')
plt.axis('off')
plt.show()

# Save the image
plt.imsave('radial_wave_image.png', myImage, cmap='gray')
```
